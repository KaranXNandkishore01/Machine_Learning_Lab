```python
import torch

# Create a sample tensor
tensor = torch.randn(3, 4)  # 3x4 matrix

# ------------------ Matrix Multiplication ------------------ #
# Method 1: Using @ operator (matrix multiplication)
y1 = tensor @ tensor.T  # Result shape: [3, 3]

# Method 2: Using matmul
y2 = tensor.matmul(tensor.T)  # Result shape: [3, 3]

# Method 3: Using matmul with out parameter
y3 = torch.empty(tensor.size(0), tensor.size(0))  # Create empty tensor with shape [3, 3]
torch.matmul(tensor, tensor.T, out=y3)

# ------------------ Element-wise Multiplication ------------------ #
# Method 1: Using * operator
z1 = tensor * tensor

# Method 2: Using mul()
z2 = tensor.mul(tensor)

# Method 3: Using mul() with out parameter
z3 = torch.empty_like(tensor)  # Same shape as original tensor [3, 4]
torch.mul(tensor, tensor, out=z3)

# Print results to verify
print("Matrix multiplication (y1):\n", y1)
print("Matrix multiplication (y2):\n", y2)
print("Matrix multiplication (y3):\n", y3)
print("Element-wise multiplication (z1):\n", z1)
print("Element-wise multiplication (z2):\n", z2)
print("Element-wise multiplication (z3):\n", z3)
```

*SVM*

```python
# Import libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train the SVM classifier
svm_model = SVC(kernel='linear')  # You can try 'rbf', 'poly', or 'sigmoid'
svm_model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = svm_model.predict(X_test)
```

```python
# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Visualize confusion matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap="Blues", fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

*Logistic Regression*

```python
# Import libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Use only class 0 and 1
X = X[y != 2]
y = y[y != 2]
```

```python
# Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train logistic regression model
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = log_model.predict(X_test)
```

```python
# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Visualize confusion matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap="Greens", fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

*Naive Baise*

```python
# Import libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Feature scaling (optional)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Create and train the Naive Bayes model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = nb_model.predict(X_test)
```

```python
# Evaluation metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Plot confusion matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap="Purples", fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Naive Bayes")
plt.show()
```
